```python

def readable_numbers(x):
    if x >= 1e6:
        s = '{:1.1f}M'.format(x * 1e-6)
    else:
        s = '{:1.0f}K'.format(x * 1e-3)
    return s
print(readable_numbers(123456))  # Example usage
```
